```python
import pandas as pd

epl_2010_11_df = pd.read_csv("epl1011leaguetable.csv")
epl_2011_12_df = pd.read_csv("epl1112leaguetable.csv")
epl_2012_13_df = pd.read_csv("epl1213leaguetable.csv")
epl_2013_14_df = pd.read_csv("epl1314leaguetable.csv")
epl_2014_15_df = pd.read_csv("epl1415leaguetable.csv")
epl_2015_16_df = pd.read_csv("epl1516leaguetable.csv")
epl_2016_17_df = pd.read_csv("epl1617leaguetable.csv")
epl_2017_18_df = pd.read_csv("epl1718leaguetable.csv")
epl_2018_19_df = pd.read_csv("epl1819leaguetable.csv")
epl_2019_20_df = pd.read_csv("epl1920leaguetable.csv")

epl_2010_11_df
```

```python
all_season = [epl_2010_11_df,
epl_2011_12_df ,
epl_2012_13_df ,
epl_2013_14_df ,
epl_2014_15_df ,
epl_2015_16_df ,
epl_2016_17_df ,
epl_2017_18_df ,
epl_2018_19_df ,
epl_2019_20_df 
]
```

```python
from functools import reduce 

all_season_df = reduce(lambda left, right:
                     pd.merge(left , right,
                              on = ["Club"],
                              suffixes=('10_11',
                                        '11_12'),
                              how = "outer"),
                     all_season)
```
